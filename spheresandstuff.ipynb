(a) Make a user defined function to plot a sphere at position x, y, z with radius, r and with any of the
available colours (ie. colour should be an input to the function). Use it to plot a single blue sphere of
radius 5 at the origin (HINT: look at matplotlib example plots)

```python
#import modules
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#setting up a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#function to plot a sphere of radius r = 5 m
#a,b,c are origin points for x,y,z axis
#r is the radius
#usercolor is the colour 
def spherePlot(a,b,c,r,usercolor):
    #usercolor = input("Pick a colour:r = red, g = green, b = blue, c = cyan, m = magenta, y = yellow, k = black, w = white ")
    
    phi = np.linspace(0.0,np.pi,100) 
    theta = np.linspace(0.0,2.0*np.pi,100)  #setting up an array for phi and theta, from 0 to pi and 2 pi 
    x = r*np.outer(np.cos(theta),np.sin(phi))+a # x co-ordinate of a sphere
    y = r*np.outer(np.sin(theta),np.sin(phi))+b # y co-ordinate of a sphere
    z = r*np.cos(phi)+c               # z co-ordinate of a shpere
    #plotting
    ax.plot_surface(x, y, z,  rstride=1, cstride=1, color=usercolor, alpha=0.6, linewidth=0)#plotting a surface with co-ordinates x,y,z

    #formating plot
    ax.set_aspect("equal") #same scaling from data to plot units for x and y
    plt.title('Sphere with radius 5 meters')

sphere = spherePlot(0,0,0,5,'b') #calling the function 
```

(b) Now write a function to calculate the energy of this array using the equation:
$$ E = −J\sum_{<ij>} S_{i}S_{j} $$

Assume J = 1. Note that the notation < ij > means that each neighbouring pair only appears once in
the sum

```python
import matplotlib.pyplot as plt #import modules
import numpy as np
#constants
n = 20 # size of the square a #initial spin array of zerosrray 20x20
J = 1 #constant given in the question

spin = np.zeros((n,n),int) #initial spin array of zeros
for i in range(n): #nested loop to assign valuese to i and j dimension
        for j in range(n): 
             if np.random.random()<0.5: #50/50 chance of a 1 or a -1 being selected (same odds)
                    spin[i,j]=1
             else:
                    spin[i,j]=-1

def energy(spin):
    I = 0 #initialisng a variable we will use
    for i in range(n-1): #summing the multiples of elements by the surrounding ones around them until the end of the x direction
        for j in range(n):
            I+=spin[i,j]*spin[i+1,j]
    
    for i in range(n): #summing the multiples of elements by the surrounding ones around them until the end of the y direction
        for j in range(n-1):
            I+=spin[i,j]*spin[i,j+1]
    
    return -J*I #multiplying the summation by the constant to give the energy

E = energy(spin)
print('The the energy is:', E,'joules')

#rint('The net magnetization is', m, 'compared to +/- 400 if they were aligned')
```

Now you will write a Metropolis algorithm Markov Chain to simulate a magnet at temperature T = 1 (also assume kB = 1). The process is as follows:


(i) Set up the systems with random spins (and calculate its energy Ei).


(ii) Choose a spin to flip at random. Calculate the energy difference (EJ − Ei).


(iii) Accept or reject the new state with probability $Pa = 1$ if $E_j < E_i$ and $Pa = \exp((E_j − E_i)/k_B T)$ if $E_j > E_i$. If you reject the new state, the step keeps the old state.

(for more on this technique see Chapter 10 of Newman’s Computational Physics or the Notebook on Monte Carlo Simulations in the Unit Github Repository)

How many steps do you need to reach equilibrium?

```python
import matplotlib.pyplot as plt #import modules
import numpy as np
np.random.seed(5)
#constants
n = 20 # size of the square a #initial spin array of zerosrray 20x20
J = 1 #constant given in the question
k_b = 1 #boltzman constant given in the question
T = 1 #contstant temp for question
beta = 1/(k_b*T) 
steps = 100000

spin = np.zeros((n,n),int) #initial spin array of zeros
for i in range(n): #nested loop to assign valuese to i and j dimension
        for j in range(n): 
             if np.random.random()<0.5: #50/50 chance of a 1 or a -1 being selected (same odds)
                    spin[i,j]=1
             else:
                    spin[i,j]=-1

def energy(spin):
    I = 0 #initialisng a variable we will use
    for i in range(n-1): #summing the multiples of elements by the surrounding ones around them until the end of the x direction
        for j in range(n):
            I+=spin[i,j]*spin[i+1,j]
    
    for i in range(n): #summing the multiples of elements by the surrounding ones around them until the end of the y direction
        for j in range(n-1):
            I+=spin[i,j]*spin[i,j+1]
    
    return -J*I #multiplying the summation by the constant to give the energy

#initilialising plots points
energyplot=[]
magplot=[]

#initialisng energy and net magnetisation
E1 = energy(spin)

for p in range(steps):
    i = np.random.randint(n) #randomly selecting an index in the i direction
    j = np.random.randint(n) #randomly selecting an index in the j direction
    
    spin[i,j] = spin[i,j] * -1 #flipping the value 
    
    E2 = energy(spin) #new energy of the new spin array
    
    D_E = E2 - E1

    if D_E > 0: #checking if E2 is bigger then E1, if so must try the acceptance condition
        if np.random.random()<np.exp(-beta*D_E): #acceptance condition
            E1 = E2 #accepted flip
        else:
            spin[i,j] = spin[i,j] * -1 #spin reverted to orinial as its not accpeted
    
    else:
        E1 = E2 # flip accepted as probability is equal to 1 (as per the question). 
       
    
    energyplot.append(E1)

plt.plot(energyplot, 'r')
plt.ylabel('Energy')
plt.xlabel('Number of steps')
plt.show
  
```

After 100 steps the system has not reached equilibrium as it still has a steep gradient (the energy is changing). It has not ran long enough for the spin to be accepted (and be aligned) to reach equilibirium(all spins are aligned so no energy can change as it wont be accepted).
to reach equilibrium you can see from the graph above that is is at about 30,000 steps, where it bottoms out at around -760.
